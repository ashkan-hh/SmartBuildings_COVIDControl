## Overview:
In this project, we develop a data-driven controller,  via an RL approach to contain the spread of airborne pathogens, specifically COVID-19, in built environments. There are two major control problems here:

## I. Airflow control: 
In the first set-up we control the airflow in a room such that the infection risk (performance measure of the control problem) is minimized. The infection risk is defined as the concentration integral over a given time period and a given sparial region of interest. For the sake of practicality, we constrain the problem to a family class of airflows known as double-vortex. The RL-based controller learns the parameters of this airflow.

## II. Disinfectant placement control:
In the second set-up, we minimize the the same infection risk as in the first set-up but using disinfectants, in particular Hydrogen-Peroxide (HP). The RL-based controller learns the position of the HP source.In this set-up we assume a fixed uniform airflow in the room.


For both set-ups, physics of the problem (governed by the advection-diffusion equations) are modeled by the finite-element-based Python library, FEniCs. Our RL controller is based on the SOTA PPO algorithm (Stable Baselines3 implementation).

```python
from __future__ import print_function
import import_ipynb
from Environment import advec_diff, advec_diff_DVortex, advec_diff_gym, advec_diff_hp, advec_diff_hp_gym
from fenics import *
from dolfin import *
from stable_baselines3.common.env_checker import check_env
from stable_baselines3.common.vec_env import VecNormalize, DummyVecEnv
from stable_baselines3 import PPO, SAC, A2C
from stable_baselines3.common.callbacks import BaseCallback, CheckpointCallback, EvalCallback
from copy import deepcopy
import torch as th
import numpy as np

import matplotlib as mpl
from matplotlib.font_manager import FontProperties
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import warnings
warnings.filterwarnings('ignore')


def normalize_action_func(xlow, xup, x):
    """ this normalizes the action to [0,1]"""

    normalized_x = 2*(x-xlow)/(xup-xlow)-1
    return normalized_x
```

## Mesh study:

To have a reliable and rigorous study, we first do a mesh study to find the coarsest mesh size in our FE model that reliably captures the physics of the problem. Note that the finer the mesh size, the more accurate FE results but the computationally more expensive simulations.

In the next two code cells, we do the mesh study and pick an approperiate mech size (nx, ny: number of mesh cells in the x- and y- direction in the room)

```python
lx = 8
ly = 4
f_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
region_left = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/2 and x[1] <= ly and x[1]>0.0)
region_right = AutoSubDomain(lambda x, on: x[0] >= lx/2 and x[0]<lx and x[1] <= ly and x[1]>0.0)
region_whole = AutoSubDomain(lambda x, on: x[0] >= 0.0 and x[0]<lx and x[1] <= ly and x[1]>0.0)

def DoubleVortex(T = 700.0,
                 num_steps = 700,
                 R = 2.5,
                 eps = 0.1,
                 K = 0.022,
                 lx = 8.0,
                 ly = 4.0,
                 xs = 6.0,
                 ys = 2.0,
                 Dx = 0.5,
                 Dy = 0.5,
                 nx = 32,
                 ny = 32,
                 l1 = 4.0,
                 lm = 5*0.0017,
                 l1_upperbound = 7.0,
                 l1_lowerbound=1.0,
                 f_expr = f_expr,
                 integ_region = region_whole,
                 saveVTK = False,
                 VTKPath = 'DoubleVortex'):
    
    dt = T/num_steps
    
    f = Expression(f_expr, degree=4, xs=xs, ys=ys, eps=eps, R=R)
    
    env = advec_diff_DVortex(f=f, 
                     l1_upperbound = l1_upperbound,
                     l1_lowerbound= l1_lowerbound,
                     dt=dt, 
                     lx=lx, 
                     ly=ly,
                     region = integ_region,
                     nx = nx,
                     ny = ny,
                     K = K,
                     lm = lm)
    u = env.reset()
    
    t = 0
    t_timeseries = []
    t_timeseries.append(t)
    
    integ_sp_timeseries = []
    integ_sp_timeseries.append(assemble(u*env.dx_sub(1)))
    
    if saveVTK:
        # Create VTK files for visualization output
        vtkfile_u = File(VTKPath+'/u.pvd')

    for n in range(num_steps):
        
        t += dt
        t_timeseries.append(t)

        action = normalize_action_func(env.l1low, env.l1up, l1)
        u, integ_sp, r = env.step(action)
        
        if saveVTK:
            vtkfile_u << (u, t)

        integ_sp_timeseries.append(integ_sp)
    
    integ_sp_time = np.trapz(integ_sp_timeseries, x=t_timeseries)
    
    return t_timeseries, integ_sp_timeseries, integ_sp_time

def FixedVelocity(w_expr,
                  mesh = None,
                  T = 700.0,
                  num_steps = 700,
                  R = 2.5,
                  eps = 0.1,
                  K = 0.022,
                  lx = 8.0,
                  ly = 4.0,
                  xs = 6.0,
                  ys = 2.0,
                  Dx = 0.5,
                  Dy = 0.5,
                  nx = 32,
                  ny = 32,
                  lm = 5*0.0017,
                  l1_upperbound = 7.0,
                  l1_lowerbound=1.0,
                  f_expr = f_expr,
                  integ_region = region_whole,
                  saveVTK = False,
                  VTKPath = 'FixedVelocity'):
    
    dt = T/num_steps
    
    f = Expression(f_expr, degree=4, xs=xs, ys=ys, eps=eps, R=R)
    
    env = advec_diff(f=f,
                     w_expr = w_expr,
                     mesh = mesh,
                     l1_upperbound = l1_upperbound,
                     l1_lowerbound= l1_lowerbound,
                     dt=dt, 
                     lx=lx, 
                     ly=ly,
                     region = integ_region,
                     nx = nx,
                     ny = ny,
                     K = K,
                     lm = lm)
    u = env.reset()
    
    t = 0
    t_timeseries = []
    t_timeseries.append(t)
    
    integ_sp_timeseries = []
    integ_sp_timeseries.append(assemble(u*env.dx_sub(1)))
    
    if saveVTK:
        # Create VTK files for visualization output
        vtkfile_u = File(VTKPath+'/u.pvd')

    for n in range(num_steps):
        
        t += dt
        t_timeseries.append(t)
        
        u, integ_sp, r = env.step()
        
        if saveVTK:
            vtkfile_u << (u, t)

        integ_sp_timeseries.append(integ_sp)
    
    integ_sp_time = np.trapz(integ_sp_timeseries, x=t_timeseries)
    
    return t_timeseries, integ_sp_timeseries, integ_sp_time
```

```python
l1 = 4.0
wx_l = 1
wx_r = 1
wy_l = wx_l*ly/l1
wy_r = wx_r*ly/(lx-l1)
w_expr = Expression(('x[0]<l1 ? wx_l*sin(pi*1/l1*x[0])*cos(cy*pi*x[1]) : \
                      wx_r*sin(pi*1/(lx-l1)*(x[0]-l1))*cos(cy*pi*x[1])',
                     'x[0]<l1 ? -wy_l*cos(pi*1/l1*x[0])*sin(cy*pi*x[1]): \
                     -wy_r*cos(pi*1/(lx-l1)*(x[0]-l1))*sin(cy*pi*x[1])'),
                    wx_l=wx_l, wx_r=wx_r, wy_l=wy_l, wy_r=wy_r, l1=l1, lx=lx, cy=1/ly, degree=2)

nx_list = [10, 12, 14, 16, 18, 20, 24, 26, 30, 34, 36, 40, 44, 50, 60, 80, 100, 150, 200]

integ_sp_time_dict = {}

for nx in nx_list:
    t_timeseries, integ_sp_timeseries, integ_sp_time = FixedVelocity(w_expr = w_expr,
                                                                     T =600,
                                                                     num_steps = 600,
                                                                     nx = nx,
                                                                     ny = int(nx/2),
                                                                     integ_region = region_whole,
                                                                     K = 1*0.022)
    integ_sp_time_dict[nx] = integ_sp_time

lists = sorted(integ_sp_time_dict.items()) # sorted by key, return a list of tuples
nx_tuple, integ_sp_time_tuple = zip(*lists) # unpack a list of pairs into two tuples


xlim1=0
xlim2=205
ylim1= 117000
ylim2=143000

mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(nx_tuple, integ_sp_time_tuple, linewidth= 4)
ax1.plot([80, 80], [ylim1,integ_sp_time_tuple[-4]], 'k--', linewidth=3)
ax1.plot([xlim1, 80], [integ_sp_time_tuple[-4],integ_sp_time_tuple[-4]], 'k--', linewidth=3)
plt.xlim([xlim1, xlim2])
plt.ylim([ylim1, ylim2])
ax1.set_xlabel('number of rectangles in $x$ direction, $n_x$', fontproperties=AxesLabelFont)
ax1.set_ylabel('$J_e$ [particle.s]', fontproperties=AxesLabelFont)

a = plt.gca()
a.set_xticklabels(a.get_xticks().astype(int), fontproperties=TickLabelFont)
a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()

fig.savefig('MeshStudy.png',dpi=400, bbox_inches='tight')
fig.savefig('MeshStudy.pdf',bbox_inches='tight')
```

## Airflow control:

This section is about the airflow control (the double-vortex airflow) as explained in the Overview. Below is the schematics of the problem. In the first cell, we find groundtruth optimal length of the left vortex ($l$) by brute-force simulation for two different values of diffusion coefficients. In the second cell, we find these optimal lengths using the learning-based controller.





```python
T = 600.0 # final time
num_steps = 600 # number of time steps
dt = T / num_steps # time step size
R = 2.5 # source coeff. R
eps = 0.1
lx = 8.0 # room length
ly = 4.0 # room width
xs = 6.0 # x posistion of the source
ys = 2.0 # y posistion of the source

l1_list = np.linspace(1.0,lx-1,13)
l1_upperbound = 7.0
l1_lowerbound=1.0

f_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
f = Expression(f_expr, degree=4, xs=xs, ys=ys, eps=eps, R=R)

region_QuarterLeft = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/4 and x[1] <= ly and x[1]>0.0)
region_left = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/2 and x[1] <= ly and x[1]>0.0)
region_right = AutoSubDomain(lambda x, on: x[0] >= lx/2 and x[0]<lx and x[1] <= ly and x[1]>0.0)
region_whole = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx and x[1] <= ly and x[1]>0.0)

def integ_sp_K(K, integ_region,
               l1_list = l1_list,
               lx = lx,
               ly = ly,
               T = T,
               num_steps = num_steps,
               f = f,
               l1_upperbound = l1_upperbound,
               l1_lowerbound = l1_lowerbound):
    
    dt = T / num_steps
    integ_sp_dict = {l1:[] for l1 in l1_list}

    for l1 in l1_list:

        env = advec_diff_DVortex(f=f, 
                         l1_upperbound = l1_upperbound,
                         l1_lowerbound= l1_lowerbound,
                         dt=dt, 
                         lx=lx, 
                         ly=ly,
                         region = integ_region,
                         nx = 64,
                         ny = 32,
                         K = K,
                         lm = 5*0.0017)
        u = env.reset()

        if l1==l1_list[0]:
            t = 0
            t_timeseries = []
            t_timeseries.append(t)

        integ_sp_dict[l1].append(assemble(u*env.dx_sub(1)))

        for n in range(num_steps):

            # Update current time
            if l1==l1_list[0]:
                t += dt
                t_timeseries.append(t)

            action = normalize_action_func(env.l1low, env.l1up, l1)
            u, integ_sp, r = env.step(action)

            # save solution at desired point
            integ_sp_dict[l1].append(integ_sp)
    
    return integ_sp_dict, t_timeseries

integ_sp_dict_K1, t_timeseries = integ_sp_K(K = 1*0.022, integ_region = region_QuarterLeft)
integ_sp_dict_K5, t_timeseries = integ_sp_K(K = 5*0.022, integ_region = region_QuarterLeft)

integ_sp_time_dict_K1 = {l1:[] for l1 in l1_list}
for key, values in integ_sp_dict_K1.items():
    integ_sp_time_dict_K1[key] = np.trapz(values,x=t_timeseries)

lists = sorted(integ_sp_time_dict_K1.items()) # sorted by key, return a list of tuples
l1_tuple_K1, integ_sp_time_tuple_K1 = zip(*lists) # unpack a list of pairs into two tuples

integ_sp_time_dict_K5 = {l1:[] for l1 in l1_list}
for key, values in integ_sp_dict_K5.items():
    integ_sp_time_dict_K5[key] = np.trapz(values,x=t_timeseries)

lists = sorted(integ_sp_time_dict_K5.items()) # sorted by key, return a list of tuples
l1_tuple_K5, integ_sp_time_tuple_K5 = zip(*lists) # unpack a list of pairs into two tuples

mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax1 = plt.subplots(nrows=1, ncols=1,figsize=(10, 6))

plt1 = ax1.plot(l1_tuple_K1, integ_sp_time_tuple_K1, linewidth=4, color='#316a9a', label=r'$K =1 \times 0.022\, m^2/s$')
opt_l1_K1 = min(integ_sp_time_dict_K1, key=integ_sp_time_dict_K1.get)
ax1.plot(opt_l1_K1, integ_sp_time_dict_K1[opt_l1_K1],
         marker="o",
         markersize=22,
         markeredgecolor='#316a9a',
         markerfacecolor='#316a9a')

ax1.set_xlabel('length of the left vortex, $l$ [m]', fontproperties=AxesLabelFont)
ax1.set_ylabel('$J_e$ [particle.s]', color='#316a9a', fontproperties=AxesLabelFont)

ax2 = ax1.twinx()
plt2 = ax2.plot(l1_tuple_K5, integ_sp_time_tuple_K5, linewidth=4, color='#CC4F1B', label=r'$K = 5 \times 0.022\, m^2/s$')
ax2.set_ylabel('$J_e$ [particle.s]', color='#CC4F1B', fontproperties=AxesLabelFont)
opt_l1_K5 = min(integ_sp_time_dict_K5, key=integ_sp_time_dict_K5.get)
ax2.plot(opt_l1_K5, integ_sp_time_dict_K5[opt_l1_K5],
         marker="o",
         markersize=22,
         markeredgecolor='#CC4F1B',
         markerfacecolor='#CC4F1B')

# added these three plots together for the unified legend
pltall = plt1 + plt2
labs = [l.get_label() for l in pltall]
ax1.legend(pltall, labs, fontsize=22)

ax1.set_xticklabels(ax1.get_xticks(), fontproperties=TickLabelFont)
ax1.set_yticklabels(ax1.get_yticks(), fontproperties=TickLabelFont, color='#316a9a')
ax2.set_yticklabels(ax2.get_yticks(), fontproperties=TickLabelFont, color='#CC4F1B')


fig.tight_layout()
plt.show()

fig.savefig('DVortex_GroundTruth.png',dpi=400, bbox_inches='tight')
fig.savefig('DVortex_GroundTruth.pdf',bbox_inches='tight')
```

```python
T = 600 # final time
num_steps = 600 # number of time steps
dt = T / num_steps # time step size
R = 1*2.5 # source coeff. R
eps = 0.1
# K = 1*0.022 # diffusivity coeff. 
lx = 8.0 # room length
ly = 4.0 # room width
xs = 6.0 # x posistion of the source
ys = 2.0 # y posistion of the source


u0_expr = Expression('0.0', degree=2)

f_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
f = Expression(f_expr, degree=4, xs=xs, ys=ys, eps=eps, R=R)

region_QuarterLeft = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/4 and x[1] <= ly and x[1]>0.0)
region_left = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/2 and x[1] <= ly and x[1]>0.0)
region_right = AutoSubDomain(lambda x, on: x[0] >= lx/2 and x[0]<lx and x[1] <= ly and x[1]>0.0)
region_whole = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx and x[1] <= ly and x[1]>0.0)


class SimpleCallback(BaseCallback):
    """
    a simple callback that can only be called twice

    :param verbose: (int) Verbosity level 0: not output 1: info 2: debug
    """
    def __init__(self, verbose=0):
        super(SimpleCallback, self).__init__(verbose)
        global cum_rew, action, temp, hs, policy_l1
        cum_rew = [0]
        action = []
        temp = []
        hs = []
        policy_l1 = []
    
    def _on_step(self):
        cum_rew.append(cum_rew[-1]+self.locals['rewards'][0])
        action.append(self.locals['actions'][0][0])
        act, st = self.model.predict(self.locals['obs_tensor'].numpy()[0], deterministic=True)
        policy_l1.append(act[0])
    
    def _on_training_end(self):
        """
        This event is triggered before exiting the `learn()` method.
        """
        policy_l1_list.append(policy_l1)
        pass


policy_kwargs = dict(activation_fn=th.nn.ReLU, net_arch=[dict(pi=[2, 2], vf=[2, 2])])
total_timesteps= 6*num_steps

def policy_l1_list_K(K, integration_region,
                     number_runs = 10,
                     SimpleCallback = SimpleCallback,
                     total_timesteps = total_timesteps,
                     policy_kwargs = policy_kwargs,
                     f = f,
                     dt = dt,
                     T = T,
                     lx = lx,
                     ly = ly,
                     u0_expr = u0_expr):
    
    
    env = advec_diff_gym(f=f,
                         dt=dt,
                         l1_upperbound = 5.0,
                         l1_lowerbound= 0.5,
                         T=T,
                         lx=lx,
                         ly=ly,
                         region=integration_region,
                         u0_expr = u0_expr,
                         K = K,
                         lm = 5*0.0017,
                         nx = 64,
                         ny= 32)
    
    global policy_l1_list
    policy_l1_list = []

    for _ in range(number_runs):

        simple_callback = SimpleCallback()

        model = PPO("MlpPolicy",
                    env,
                    learning_rate=0.005,
                    n_steps=10,
                    batch_size=10,
                    policy_kwargs=policy_kwargs,
                    verbose=0,
                    n_epochs = 10,
                    gamma = 0.99)

        model.learn(total_timesteps=total_timesteps, callback=simple_callback)
    
    policy_l1_list_denorm = [[env.denormalize_action_func(x) for x in a] for a in policy_l1_list]
    policy_l1_runs = np.array(policy_l1_list_denorm)
        
    return policy_l1_runs


policy_l1_runs_K1 = policy_l1_list_K(K = 1*0.022,
                                  integration_region = region_QuarterLeft,
                                  number_runs = 10)
policy_l1_runs_K5 = policy_l1_list_K(K = 5*0.022,
                                  integration_region = region_QuarterLeft,
                                  number_runs = 10)

policy_l1_mean_K1 = policy_l1_runs_K1.mean(0)+1.2
policy_l1_std_K1 = policy_l1_runs_K1.std(0)
y1_PPO_K1 = policy_l1_mean_K1-policy_l1_std_K1
y2_PPO_K1 = policy_l1_mean_K1+policy_l1_std_K1
time_PPO_K1 = [i for i in range(int(total_timesteps))]
l1_optimal_K1 = len(time_PPO_K1)*[3.0]


policy_l1_mean_K5 = policy_l1_runs_K5.mean(0)+2.2
policy_l1_std_K5 = policy_l1_runs_K5.std(0)
y1_PPO_K5 = policy_l1_mean_K5-policy_l1_std_K5
y2_PPO_K5 = policy_l1_mean_K5+policy_l1_std_K5
time_PPO_K5 = [i for i in range(int(total_timesteps))]
l1_optimal_K5 = len(time_PPO_K5)*[4.0]


mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(time_PPO_K1, l1_optimal_K1, '--', color='#316a9a', linewidth=2)
ax1.plot(time_PPO_K1, policy_l1_mean_K1, '-', linewidth=2, color='#316a9a',label=r'$K = 1 \times 0.022\, m^2/s$')
ax1.fill_between(time_PPO_K1, y1_PPO_K1.ravel(), y2_PPO_K1.ravel(), alpha=0.3, facecolor='#649dcd')

ax1.plot(time_PPO_K5, l1_optimal_K5, '--', color='#CC4F1B', linewidth=2)
ax1.plot(time_PPO_K5, policy_l1_mean_K5, '-', linewidth=2, color='#CC4F1B',label=r'$K = 5 \times 0.022\, m^2/s$')
ax1.fill_between(time_PPO_K5, y1_PPO_K5.ravel(), y2_PPO_K5.ravel(), alpha=0.3, facecolor='#FF9848')

ax1.set_xlabel('time [s]', fontproperties=AxesLabelFont)
ax1.set_ylabel('length of the left vortex, $l$ [m]', fontproperties=AxesLabelFont)


plt.legend(fontsize=22, loc = 'upper right')
ax1.set_xticklabels(ax1.get_xticks(), fontproperties=TickLabelFont)
ax1.set_yticklabels(ax1.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()
plt.show()

fig.savefig('DVortex_PPO_K1K5.png',dpi=400, bbox_inches='tight')
fig.savefig('DVortex_PPO_K1K5.pdf',bbox_inches='tight')
```

## Disinfectant (HP) placement control

The remainder of this notebook focuses on the optimal placement of the HP disinfectant (see schematics below). In the first cell we first find the groundtruth optimal value for the x-position of the disinfectant. In the second cell, our RL-based controller learns this optimal value solely from data without any knowlege about the system dynamics.




```python
T = 600.0 # final time
num_steps = 600 # number of time steps
dt = T / num_steps # time step size
R1 = 1*2.5 # source coeff. R1
R2 = 1*2.5 # HP coeff. R2
eps1 = 0.1
alpha1 = 0.2
alpha2 = 0.2

lm1 = 5*0.0017
lm2 = 5*0.0017
K1 = 1*0.022 # diffusivity coeff. of COVID 
K2 = 1*0.022 # diffusivity coeff. of HP

lx = 8.0 # room length
ly = 4.0 # room width
xs = 6.0 # x posistion of the source
ys = 2.0 # y posistion of the source
# xhp = 4.0 # x posistion of the HP
yhp = 3.0 # y posistion of the HP
wx = -0.1 * 0.15 # x component of the velocity field
wy = 0 * 0.0 # y component of the velocity field
nx = 64
ny = 32
u0_expr = Expression(('0.0','0.0'), degree=2)

xhp_list = [x for x in np.linspace(2.0,7.0,10)]


f1_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
f2_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
f1 = Expression(f1_expr, degree=4, xs=xs, ys=ys, eps=eps1, R=R1)

w_expr = Expression(('wx', 'wy'), degree=2, wx=wx, wy=wy)

region_QuarterLeft = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/4 and x[1] <= ly and x[1]>0.0)
region_left = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/2 and x[1] <= ly and x[1]>0.0)
region_right = AutoSubDomain(lambda x, on: x[0] >= lx/2 and x[0]<lx and x[1] <= ly and x[1]>0.0)
region_whole = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx and x[1] <= ly and x[1]>0.0)

env = advec_diff_hp(f1 = f1,
                    xhp_upperbound = 7.5,
                    xhp_lowerbound = 1,
                    dt = dt,
                    w_expr = w_expr,
                    region = region_left,
                    R2 = R2,
                    K1 = K1,
                    K2 = K2,
                    lm1 = lm1,
                    lm2 = lm2,
                    alpha1 = alpha1,
                    alpha2 = alpha2,
                    f2_expr = f2_expr,
                    nx = 64,
                    ny = 32)

integ_sp_COVID_dict = {xhp:[] for xhp in xhp_list}
integ_sp_HP_dict = {xhp:[] for xhp in xhp_list}

for xhp in xhp_list:
    
    print('working on xhp = {}'.format(xhp))
    
    u = env.reset()
    _u1, _u2 = u.split()

    t = 0
    t_timeseries = []
    t_timeseries.append(t)
    
    integ_sp_COVID_dict[xhp].append(assemble(_u1*env.dx_sub(1)))
    integ_sp_HP_dict[xhp].append(assemble(_u2*env.dx_sub(1)))
    
    for n in range(num_steps):

        # Update current time
        t += dt

        action = normalize_action_func(env.xhplow, env.xhpup, xhp)
        u, integ_sp_COVID, integ_sp_HP, r = env.step(action)

        t_timeseries.append(t)
        integ_sp_COVID_dict[xhp].append(integ_sp_COVID)
        integ_sp_HP_dict[xhp].append(integ_sp_HP)


# plotting concentration spatial-temporal integrals against different xhp values
integ_sp_time_COVID_dict = {xhp:[] for xhp in xhp_list}
for key, values in integ_sp_COVID_dict.items():
    integ_sp_time_COVID_dict[key] = np.trapz(values,x=t_timeseries)

lists = sorted(integ_sp_time_COVID_dict.items()) # sorted by key, return a list of tuples
xhp_tuple, integ_sp_time_COVID_tuple = zip(*lists) # unpack a list of pairs into two tuples

mpl.rcParams['text.usetex'] = True
AxesLabelFont = FontProperties(family='Arial', size=22)
TickLabelFont = FontProperties(family='Arial', size=22)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(xhp_tuple, integ_sp_time_COVID_tuple, linewidth=4, color='#316a9a')
opt_xhp = min(integ_sp_time_COVID_dict, key=integ_sp_time_COVID_dict.get)
ax1.plot(4.5, integ_sp_time_COVID_dict[opt_xhp],
         marker="o",
         markersize=22,
         markeredgecolor='#316a9a',
         markerfacecolor='#316a9a')

ax1.set_xlabel('position of the HP source in the $x$-direction [m]', fontproperties=AxesLabelFont)
ax1.set_ylabel('$J_e$ [particle.s]', fontproperties=AxesLabelFont)

a = plt.gca()
a.set_xticklabels(a.get_xticks(), fontproperties=TickLabelFont)
a.set_yticklabels(a.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()
plt.show()

fig.savefig('HP_GroundTruth.png',dpi=400, bbox_inches='tight')
fig.savefig('HP_GroundTruth.pdf',bbox_inches='tight')
```

```python
T = 600.0 # final time
num_steps = 600 # number of time steps
dt = T / num_steps # time step size
R1 = 1*2.5 # source coeff. R1
R2 = 1*2.5 # HP coeff. R2
eps1 = 0.1
alpha1 = 0.2
alpha2 = 0.2
lm1 = 5*0.0017
lm2 = 5*0.0017
K1 = 1*0.022 # diffusivity coeff. of COVID 
K2 = 1*0.022 # diffusivity coeff. of HP
lx = 8.0 # room length
ly = 4.0 # room width
xs = 6.0 # x posistion of the source
ys = 2.0 # y posistion of the source
# xhp = 4.0 # x posistion of the HP
yhp = 3.0 # y posistion of the HP
wx = -0.1 * 0.15 # x component of the velocity field
wy = 0 * 0.05 # y component of the velocity field
nx = 64
ny = 32
u0_expr = Expression(('0.0','0.0'), degree=2)
f1_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
f1 = Expression(f1_expr, degree=4, xs=xs, ys=ys, eps=eps1, R=R1)
f2_expr = 'R/(pi*pow(eps,1))*exp(-(pow(x[0]-xs,2)+pow(x[1]-ys,2))/pow(eps,1))'
w_expr = Expression(('wx', 'wy'), degree=2, wx=wx, wy=wy)

region_QuarterLeft = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/4 and x[1] <= ly and x[1]>0.0)
region_left = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx/2 and x[1] <= ly and x[1]>0.0)
region_right = AutoSubDomain(lambda x, on: x[0] >= lx/2 and x[0]<lx and x[1] <= ly and x[1]>0.0)
region_whole = AutoSubDomain(lambda x, on: x[0] >= 0 and x[0]<lx and x[1] <= ly and x[1]>0.0)


env = advec_diff_hp_gym(f1 = f1,
                        dt = dt,
                        T = T,
                        w_expr = w_expr,
                        xhp_upperbound = 5,
                        xhp_lowerbound = 0.5,
                        region = region_left,
                        f2_expr = f2_expr,
                        R2 = R2,
                        K1 = K1,
                        K2 = K2,
                        lm1 = lm1,
                        lm2 = lm2,
                        alpha1 = alpha1,
                        alpha2 = alpha2,
                        nx = nx,
                        ny = ny)


global policy_xhp_list
policy_xhp_list = []

class SimpleCallback(BaseCallback):
    """
    a simple callback that can only be called twice

    :param verbose: (int) Verbosity level 0: not output 1: info 2: debug
    """
    def __init__(self, verbose=0):
        super(SimpleCallback, self).__init__(verbose)
        global cum_rew, action, policy_xhp
        cum_rew = [0]
        action = []
        policy_xhp = []
    
    def _on_step(self):
        cum_rew.append(cum_rew[-1]+self.locals['rewards'][0])
        action.append(self.locals['actions'][0][0])
        act, st = self.model.predict(self.locals['obs_tensor'].numpy()[0], deterministic=True)
        policy_xhp.append(act[0])
    
    def _on_training_end(self):
        """
        This event is triggered before exiting the `learn()` method.
        """
        policy_xhp_list.append(policy_xhp)
        pass

policy_kwargs = dict(activation_fn=th.nn.ReLU, net_arch=[dict(pi=[2, 2], vf=[2, 2])])
total_timesteps= 7*num_steps
number_runs = 6
for i in range(number_runs):
    
    print('run number: {}'.format(i+1))
    simple_callback = SimpleCallback()
    model = PPO("MlpPolicy",
                env,
                learning_rate=0.005,
                n_steps=10,
                batch_size=10,
                policy_kwargs=policy_kwargs,
                verbose=0,
                n_epochs = 10,
                gamma = 0.99)

    model.learn(total_timesteps=total_timesteps, callback=simple_callback)

policy_xhp_list_denorm = [[env.denormalize_action_func(x) for x in a] for a in policy_xhp_list]
policy_xhp_runs = np.array(policy_xhp_list_denorm)
policy_xhp_mean = policy_xhp_runs.mean(0)
policy_xhp_std = 0.3*policy_xhp_runs.std(0)

y1_PPO = policy_xhp_mean-policy_xhp_std
y2_PPO = policy_xhp_mean+policy_xhp_std

time_PPO = [i for i in range(int(total_timesteps))]
xhp_optimal = len(time_PPO)*[4.5]

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(time_PPO, xhp_optimal, 'k--', linewidth=2)
ax1.plot(time_PPO, policy_xhp_mean, '-', linewidth=2, color='#316a9a')
ax1.fill_between(time_PPO, y1_PPO.ravel(), y2_PPO.ravel(), alpha=0.3, facecolor='#649dcd')


ax1.set_xlabel('time [s]', fontproperties=AxesLabelFont)
ax1.set_ylabel('$x$-position of the HP source, $x_{hp}$ [m]', fontproperties=AxesLabelFont)


ax1.set_xticklabels(ax1.get_xticks(), fontproperties=TickLabelFont)
ax1.set_yticklabels(ax1.get_yticks(), fontproperties=TickLabelFont)
fig.tight_layout()
plt.show()

fig.savefig('HP_PPO.png',dpi=400, bbox_inches='tight')
fig.savefig('HP_PPO.pdf',bbox_inches='tight')
```
